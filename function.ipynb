Задание 1


```python
import pandas as pd

# Чтение данных из файлов
movies_df = pd.read_csv("D:\\movies.csv")
ratings_df = pd.read_csv("D:\\ratings.csv")

# Вычисление средних оценок для каждого фильма
average_ratings = ratings_df.groupby("movieId")["rating"].mean().reset_index()

# Объединение данных по полю "movieId" с учетом средних оценок
merged_df = pd.merge(average_ratings, movies_df, on="movieId")

# Функция классификации фильмов
def classify_rating(rating):
    if rating <= 2:
        return "Низкий рейтинг"
    elif rating <= 4:
        return "Средний рейтинг"
    else:
        return "Высокий рейтинг"

# Применение функции классификации, создание нового столбца "class"
merged_df["class"] = merged_df["rating"].apply(classify_rating)

# Вывод результата с форматированием до десятых для удобства просмотра 
merged_df["rating"] = merged_df["rating"].round(1)  # Округляем оценку до десятых
pd.set_option("display.float_format", "{:.1f}".format)
print(merged_df[["title", "rating", "class"]])

```

Задание 2. Вывод результатов запросов, где определен регион.

```python
import pandas as pd

# Чтение данных из файла keywords.csv
keywords_df = pd.read_csv("D:\\keywords.csv")

# Словарь с географическими данными
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

# Функция для классификации по регионам
def classify_region(keyword):
    for region, cities in geo_data.items():
        for city in cities:
            if city in keyword:
                return region
    return None

# Применение функции к столбцу 'keyword'
keywords_df['region'] = keywords_df['keyword'].apply(classify_region)

# Удаление строк с пропущенными значениями в столбце 'region'
result_df = keywords_df.dropna(subset=['region'])

# Вывод результата
print(result_df[['keyword', 'region']])
```
